```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None) 
```

```python
df1 = pd.read_csv("orders-2024-11-09-19-45-10.csv")
df2 = pd.read_csv("orders-2024-11-09-19-45-26.csv")
```

```python
print(df1.shape)
df1.head()
```

```python
print(df2.shape)
df2.head()
```

```python
sales_data = pd.concat([df2,df1], ignore_index = True)
print(sales_data.shape)
print(sales_data.head())
```

```python
logistic_data = pd.read_csv("LogisticsCheckAccounts_20241009085843.csv",dtype=str, index_col=False)

print(logistic_data.shape)
logistic_data.head()
```

```python
sales_data = sales_data.dropna(subset = ["廠商交易編號"])
```

```python
sales_data = sales_data.rename(columns={'廠商交易編號': '廠商訂單編號'})
sales_data.info()
```

```python
merged_data = pd.merge(sales_data,logistic_data, on = "廠商訂單編號", how = "inner")
merged_data.head()
```

```python
invoice = pd.read_csv("invoice.csv",encoding='big5')
invoice = invoice.loc[[], :]
invoice
```

```python
invoice.info()
```

```python
def add_purchase_details(num, import_number):
    items_raw= merged_data.loc[merged_data['廠商訂單編號'] == num,'商品名稱'].values[0]
    amounts_raw = merged_data.loc[merged_data['廠商訂單編號'] == num,'數量'].values[0]

    items = str(items_raw).split('\r\n')
    amounts = str(amounts_raw).split('\r\n')   
   
    merged_details = dict(zip(items, amounts))
    merged_details['運費'] = 1
    add_row = pd.DataFrame(columns=['匯入單號','商品名稱','購買數量'])

    for key, value in merged_details.items():
        new_row = {'匯入單號': import_number, '商品名稱': key, '購買數量': value}
        add_row = pd.concat([add_row, pd.DataFrame([new_row])], ignore_index=True)
    
    return add_row
```

```python
index = 0
import_number = 0
for num in merged_data['廠商訂單編號']:
    import_number += 1
    invoice.loc[index,'自訂編號'] = num
    invoice.loc[index,'客戶名稱'] = merged_data.loc[merged_data['廠商訂單編號'] == num,'姓名'].values
    invoice.loc[index,'客戶電子信箱'] = merged_data.loc[merged_data['廠商訂單編號'] == num,'Email'].values
    invoice.loc[index,'課稅別'] = 1
    invoice.loc[index,'發票金額'] = merged_data.loc[merged_data['廠商訂單編號'] == num,'訂單金額'].values
    invoice.loc[index,'發票捐贈'] = 0
    invoice.loc[index,'載具類別'] = 1
    invoice.loc[index,'列印註記'] = 0
    invoice.loc[index,'字軌類別'] = int(7)
    invoice.loc[index,'匯入單號'] = int(import_number)

    add_row = add_purchase_details(num,import_number)
    invoice.loc[index,'商品名稱'] = add_row.loc[0,'商品名稱']
    invoice.loc[index,'購買數量'] = add_row.loc[0,'購買數量']
    invoice = pd.concat([invoice, add_row.iloc[1:]], ignore_index=True)

    index = len(invoice) + 1
    
```

```python
col = ['匯入單號','課稅別','發票捐贈','載具類別','列印註記','字軌類別']
invoice[col].astype('Int64')
invoice.to_csv('output.csv', index=False, encoding='utf-8')
```
